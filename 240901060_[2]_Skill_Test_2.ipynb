```python
import pandas as pd

# Load the dataset
file_path = "/content/student_scores.csv"
df = pd.read_csv(file_path)

# Display first few rows
print("Initial Data:")
print(df.head())

# -----------------------------
# a. Summary statistics
# -----------------------------
print("\nSummary Statistics for Each Subject:")
print(df[['Maths', 'Physics', 'Chemistry']].describe())

# -----------------------------
# b. Average score per student
# -----------------------------
df['Average Score'] = df[['Maths', 'Physics', 'Chemistry']].mean(axis=1)

print("\nAverage Score for Each Student:")
print(df[['Student ID', 'Name', 'Average Score']])

# -----------------------------
# c. Students scoring below 60 in more than two subjects
# -----------------------------
# Count number of subjects where score < 60
df['Below_60_Count'] = (df[['Maths', 'Physics', 'Chemistry']] < 60).sum(axis=1)

# Filter students with more than 2 subjects below 60
low_performers = df[df['Below_60_Count'] > 2]

print("\nStudents scoring below 60 in more than two subjects:")
if low_performers.empty:
    print("No students found.")
else:
    print(low_performers[['Student ID', 'Name', 'Maths', 'Physics', 'Chemistry']])
```
