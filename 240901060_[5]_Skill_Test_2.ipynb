```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create date range for one year
dates = pd.date_range(start='2023-01-01', end='2023-12-31')

# Step 2: Generate random temperature data using NumPy
np.random.seed(0)  # for reproducibility
temperatures = 20 + 10 * np.sin(np.linspace(0, 3 * np.pi, len(dates))) + np.random.normal(0, 2, len(dates))

# Step 3: Create pandas DataFrame
df = pd.DataFrame({'Date': dates, 'Temperature': temperatures})

# Step 4: Set Date column as index
df.set_index('Date', inplace=True)

# Step 5: Display first few rows
print("First 5 rows:")
print(df.head())

# Step 6: Basic statistics
print("\nStatistics:")
print(df.describe())

# Step 7: Calculate rolling average (7-day moving average)
df['Rolling_Avg'] = df['Temperature'].rolling(window=7).mean()

# Step 8: Find hottest and coldest day
max_temp = df['Temperature'].max()
min_temp = df['Temperature'].min()

print("\nMax Temperature:", max_temp)
print("Min Temperature:", min_temp)

# Step 9: Plot the data
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Temperature'], label='Daily Temperature', alpha=0.7)
plt.plot(df.index, df['Rolling_Avg'], label='7-Day Rolling Average', color='red')

# Step 10: Add labels and title
plt.title('Daily Temperature Time Series (2023)')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()

# Step 11: Show plot
plt.show()
```
