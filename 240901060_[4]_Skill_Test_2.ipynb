```python
import math

# 1. Euclidean Distance for two points in 2D
def euclidean_distance(x1, y1, x2, y2):
    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return distance

# 2. Dot Product of two vectors
def dot_product_vectors(v1, v2):
    if len(v1) != len(v2):
        return "Vectors must be of same length"

    dot_product = sum(v1[i] * v2[i] for i in range(len(v1)))
    return dot_product

# 3. Dot Product of two matrices (matrix multiplication)
def dot_product_matrices(A, B):
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]

    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]

    return result

# ---- Example Usage ----

# Euclidean Distance
x1, y1 = 1, 2
x2, y2 = 4, 6
print("Euclidean Distance:", euclidean_distance(x1, y1, x2, y2))

# Dot Product (Vectors)
v1 = [1, 2, 3]
v2 = [4, 5, 6]
print("Dot Product (Vectors):", dot_product_vectors(v1, v2))

# Dot Product (Matrices)
A = [[1, 2],
     [3, 4]]

B = [[5, 6],
     [7, 8]]

print("Dot Product (Matrices):")
for row in dot_product_matrices(A, B):
    print(row)
```
